```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv'
titanic = pd.read_csv(url)
```

**2.Feature names with null values**

```python
null_columns = titanic.columns[titanic.isnull().any()]
print("Columns with null values:", null_columns.tolist())
```

**3.Filling Missing Values**

```python
titanic['Age'].fillna(titanic['Age'].mean())
#df.method({Age: mean()}, inplace=True)

# Fill missing 'Cabin' values with 'Unknown'
titanic['Cabin'].fillna('Unknown')
```

**4.Plotting Dtaset with Two  Bars**

```python
titanic['Survival Status'] = titanic['Survived'].map({0: 'Dead', 1: 'Survived'})

# Plot the data
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Survival Status', hue='Sex', palette=['#1f77b4', '#ff7f0e'])
plt.title('Survival Count by Gender')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()
```

**5.Survival Counter By Passenger class**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Survival Status', hue='Pclass', palette='viridis')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.legend(title='Passenger Class')
plt.show()
```

**6.Create a column named 'AgeClass' and insert value**

```python
bins = [0, 16, 26, 36, 62, float('inf')]
labels = [0, 1, 2, 3, 4]

# Create 'AgeClass' column
titanic['AgeClass'] = pd.cut(titanic['Age'], bins=bins, labels=labels, right=False)
```
