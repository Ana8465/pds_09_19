```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = 'https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv'
titanic = pd.read_csv(url)
```

```python
titanic['Age'].fillna(titanic['Age'].mean() )
#df.method({Age: mean()}, inplace=True)

# Fill missing 'Cabin' values with 'Unknown'
titanic['Cabin'].fillna('Unknown')
```

```python
titanic['Survival Status'] = titanic['Survived'].map({0: 'Dead', 1: 'Survived'})

# Plot the data
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic, x='Survival Status', hue='Sex', palette=['#1f77b4', '#ff7f0e'])
plt.title('Survival Count by Gender')
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.show()
```

```python
bins = [0, 16, 26, 36, 62, float('inf')]
labels = [0, 1, 2, 3, 4]

# Create 'AgeClass' column
titanic['AgeClass'] = pd.cut(titanic['Age'], bins=bins, labels=labels, right=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter only survivors
survivors = titanic[titanic['Survived'] == 1]

# Plot scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for Male survivors
plt.scatter(
    survivors[survivors['Sex'] == 'male']['Age'],
    survivors[survivors['Sex'] == 'male']['Fare'],
    color='green', label='Male', alpha=0.6
)

# Scatter plot for Female survivors
plt.scatter(
    survivors[survivors['Sex'] == 'female']['Age'],
    survivors[survivors['Sex'] == 'female']['Fare'],
    color='red', label='Female', alpha=0.6
)

# Labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Male and Female Survivors')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
