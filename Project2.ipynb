```python

import csv
from google.colab import files
import pandas as pd
import random

header1 = ['Registration Number', 'Name', 'TT-1 Marks']


data1 = [
    [f'2025REG{i:03}', f'Student {i}', random.randint(0, 20)]
    for i in range(1, 51)
]

# Write data to term-test-1 CSV file
with open('term-test-1-result.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header1)
    writer.writerows(data1)

# Verify the file is saved
!ls -lh term-test-1-result.csv
#files.download('term-test-1-result.csv')


# Define the header for term-test-2
header2 = ['Registration Number', 'Name', 'TT-2 Marks']

# Generate sample data for 50 students for term-test-2
# (using the same registration numbers and names as term-test-1)
data2 = [
    [f'2025REG{i:03}', f'Student {i}', random.randint(0, 20)]
    for i in range(1, 51)
]

# Write data to term-test-2 CSV file
with open('term-test-2-result.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header2)
    writer.writerows(data2)

# Verify the file is saved
!ls -lh term-test-2-result.csv
#files.download('term-test-2-result.csv')

# Load the CSV files into pandas DataFrames
df1 = pd.read_csv('term-test-1-result.csv')
df2 = pd.read_csv('term-test-2-result.csv')

# Merge the two dataframes based on 'Registration Number'
merged_df = pd.merge(df1, df2, on='Registration Number')

# Display the merged dataframe
merged_df
```

```python
attendance_term_final_data = []
for student in data:
    reg_num = student[0]
    attendance_marks = random.randint(0, 10)  # Random attendance marks between 0 and 10
    term_final_marks = random.randint(0, 100)  # Random term final marks between 0 and 100
    attendance_term_final_data.append([reg_num, attendance_marks, term_final_marks])

# Write to CSV
attendance_header = ['Registration Number', 'Attendance Marks', 'Term Final Marks']
with open('attendance_term_final.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(attendance_header)
    writer.writerows(attendance_term_final_data)

# Load the new CSV into a DataFrame
df_attendance_final = pd.read_csv('attendance_term_final.csv')

# Merge with the existing DataFrame
final_df = pd.merge(merged_df, df_attendance_final, on='Registration Number')

# Calculate 'Final Marks'
final_df['Final Marks'] = final_df['Term Final Marks'] * 0.7 + final_df['Average TT Marks'] + final_df['Attendance Marks']

# Display the final DataFrame
print(final_df.head())
```

```python
# Write the final DataFrame to CSV
final_df.to_csv('final_result.csv', index=False)
```
